```python
#Visuallization
import pandas
import matplotlib.pyplot as plt
```

```python
df=pandas.read_csv('tips.csv')
df.head()
```

```python
#scatter plot
plt.scatter(df['day'],df['tip'])
plt.title('scatter plot')
plt.show()
```

```python
#linear plot
plt.plot(df['tip'])
plt.plot(df['size'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
#bar plot
plt.bar(df['day'],df['tip'])
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
```

```python
#histogram
plt.hist(df['total_bill'])
plt.title("Histogram")
plt.show()
```

```python
import seaborn as sns
#linear plot
sns.lineplot(x="sex",y="total_bill",data=df)
#scatter plot
sns.scatterplot(x='day',y='tip',data=df,hue='sex')
#bar plot
sns.barplot(x='day',y='tip',data=df,hue='sex')
#histogram
sns.histplot(x='total_bill',data=df,kde=True,hue='sex')
plt.show()
```
